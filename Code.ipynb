```python
import pandas as pd
import glob
df_name = pd.read_csv('symbols_valid_meta.csv')
path = 'etfs'
all_files = glob.glob(path + "/*.csv")
df = pd.DataFrame()
for file in all_files:
    print(file)
    df_temp = pd.read_csv(file)
    start = file.index('\\') + 1
    end = file.index('.')
    symbol = file[start:end]
    df_temp['Symbol'] = symbol
    # df_temp['Security Name'] = df_name[df_name['Symbol'] == symbol]['Security Name']
    df = pd.concat([df, df_temp])
```

# 2

```python
window_size = 30 #30 days
rolling_mean = df_final.groupby('Symbol')[['Volume']].rolling(window_size).mean()
print(rolling_mean)
rolling_median = df_final.groupby('Symbol')[['Volume']].rolling(window_size).median()
print(rolling_median)
```

```python
df_final['vol_moving_avg'] = rolling_mean['Volume'].tolist()
df_final['adj_close_rolling_med'] = rolling_median['Volume'].tolist()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
df_final.dropna(inplace=True)
print(1)
# Select features and target
features = ['vol_moving_avg', 'adj_close_rolling_med']
target = 'Volume'
print(2)
X = df_final[features]
y = df_final[target]
print(3)
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(4)
# Create a RandomForestRegressor model
model = RandomForestRegressor(n_estimators=100, random_state=42)
print(5)
```

```python
# Train the model
model = RandomForestRegressor(n_estimators=10, random_state=42)
model.fit(X_train, y_train)
```

```python
print(6)
# Make predictions on test data
y_pred = model.predict(X_test)
print(7)
# Calculate the Mean Absolute Error and Mean Squared Error
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
```
